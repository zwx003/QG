```python
X=bt.data
y=bt.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(bt.data, bt.target, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
std_X = StandardScaler()
```

```python
X_train = std_X.fit_transform(X_train)
X_test = std_X.fit_transform(X_test)
```

```python
std_y = StandardScaler()
```

```python
y_train = std_y.fit_transform(y_train.reshape(-1, 1))  # æ•°æ®ä¸ºäºŒç»´
y_test = std_y.fit_transform(y_test.reshape(-1, 1))
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor, Ridge
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
```

```python
y_lr_predict = std_y.inverse_transform(lr.predict(X_test))
```

```python
from sklearn.metrics import mean_squared_error
print('å‡æ–¹è¯¯å·®ï¼š', mean_squared_error(std_y.inverse_transform(y_test), y_lr_predict))
print(f'R : {lr.score(X_test, y_test)}')
```

```python
sgd = SGDRegressor()
sgd.fit(X_train, y_train)
```

```python
y_sgd_predict = std_y.inverse_transform(sgd.predict(X_test))
print(y_sgd_predict)
print("å‡æ–¹è¯¯å·®ï¼š", mean_squared_error(std_y.inverse_transform(y_test), y_sgd_predict))
print(f"R : {sgd.score(X_test, y_test)}")
```
